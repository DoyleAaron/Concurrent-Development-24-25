```python
import pandas
import matplotlib.pyplot as plt

df1 = pandas.read_csv('fps_log.csv')
df2 = pandas.read_csv('2_thread_fps_log.csv')
df3 = pandas.read_csv('4_thread_fps_log.csv')
df4 = pandas.read_csv('8_thread_fps_log.csv')
```

These are just the basic imports that I'm using to create the graph and I'm reading in the CSV file

```python
plt.plot(df1['Timestamp'], df1['FPS'], label='1 thread')
plt.plot(df2['Timestamp'], df2['FPS'], label='2 threads')
plt.plot(df3['Timestamp'], df3['FPS'], label='4 threads')
plt.plot(df4['Timestamp'], df4['FPS'], label='8 threads')
plt.legend()   
plt.title('FPS recorded per second')
plt.xlabel('Second')
plt.ylabel('FPS')
plt.show() 
```

## Graph Description
This is the graph that I have created showing the fps per second whilst running this simulation. This is only for single threaded currently and I will add more as I create them.

As you can see in this graph the frames per second does seem to fluctatuate quite a bit but I believe this is due to when the fish are getting their big bursts of populating as they are so collected and congregated it is causing this frame drop

## Two Threads Analysis
As you can see on the graph above, both simulations are quite chaotic and can range depending on the simulation but overall the two threaded version runs a little bit slower than the single threaded version, I beielve this is becasue theres is more time spent using the waitgroups and mutex locks. As it is a small program this makes sense but if you were to use more threads on a larger project it would more than  likely split them up, just in this scenario where they don't have the most taxing workload it doesn't make a big difference.

## 4 Threads Analysis
Interestingly, 4 threads overall runs better than the other two, I believe this is beacuse most CPUs have 4 cores, it can all run concurrently and make the most of the CPU allowing for everything to run that little bit faster. It also makes sense as each thread has to deal with a lesser amount of work maiking it run quicker overall.

## 8 Threads Analysis
When I moved the simulation to using 8 threads instead of 4, it became a lot more consistent, it for the most part idn't have as high highs or low lows, I believe this is because my macbook has an 8 core CPU which is allowing it to be more stable as all cores can be utilised by a single thread
